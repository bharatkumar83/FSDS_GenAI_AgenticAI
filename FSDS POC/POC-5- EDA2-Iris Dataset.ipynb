### Visuvalizing the dataset

#### Barplot

```python
sns.countplot(x= 'Species', hue='Species',data=iris) # plotiing barplot for the Species attribute
plt.show()
```

```python
iris.head()
```

#### joint plot

```python
fig= sns.jointplot(x='SepalLengthCm',y='SepalWidthCm',data=iris) # ploting the joint plot for the attributtes SepalLength and Sepalwidth
```

```python
fig1= sns.jointplot(x='SepalLengthCm',y='SepalWidthCm',data=iris,kind= 'reg') # jointplot with regression line
```

```python
fig2= sns.jointplot(x='SepalLengthCm',y='SepalWidthCm',data=iris,kind= 'hex') # hexagonal joint plot
```

#### FacetGrid Plot

```python
import matplotlib.pyplot as plt  # scatter plot uing Sepallength and sepalwidth colored with Species
%matplotlib inline

sns.FacetGrid(data=iris,hue='Species',height=5).map(plt.scatter,'SepalLengthCm','SepalWidthCm').add_legend()
```

```python
iris.head()
```

#### Boxplot  or  Whisker plot

```python
fig= plt.gcf()   # gcf means get current figure used to when no fig the  automatically create a new one.
fig.set_size_inches(10,7) # using the boxplot on species and petallength
fig= sns.boxplot(x='Species', y='PetalLengthCm', data= iris, hue='Species',order=['Iris-virginica','Iris-versicolor','Iris-setosa'], linewidth=2.5,orient='v',dodge=False)
```

```python
iris.boxplot(by='Species', figsize=(12,6)) #  ploting the box plot grouped by species
```

#### Strip plot

```python
fig = plt.gcf()
fig.set_size_inches(10,7) # strip plot is used to displays individual data points along a single axis
fig= sns.stripplot(x='Species', y='SepalLengthCm',data= iris, jitter= True, edgecolor='gray',size=8, palette='winter',orient='v')
```

#### Combining Boxplot and Strip plots

```python
fig = plt.gcf() 
fig.set_size_inches(10,7)
fig= sns.boxplot(x='Species',y='SepalLengthCm',data= iris, hue='Species') # combination of Boxplot and strip plot
fig= sns.stripplot(x='Species',y='SepalLengthCm', data= iris, hue='Species', jitter= True, edgecolor='gray')
```

```python
ax= sns.boxplot(x='Species',y='PetalLengthCm',data= iris)
ax=sns.stripplot(x='Species',y='PetalLengthCm',data= iris,hue= "Species", jitter= True,edgecolor='gray')

boxone=ax.patches[0]
boxone.set_facecolor('green') # using the patches to the individual box plot which is used to make changes in the plot
boxone.set_edgecolor('black')

boxtwo= ax.patches[1]
boxtwo.set_facecolor('Yellow')
boxtwo.set_edgecolor('black')

boxthree= ax.patches[2]
boxthree.set_facecolor('red')
boxthree.set_edgecolor('black')
```

#### Violin plot

#### Heatmap

```python
num_iris=iris.select_dtypes(include='number') # creating a data frame for numerical values 
```

```python
fig= plt.gcf()
fig.set_size_inches(10,7) # Correlation b/w the variables in the dataframe
fig= sns.heatmap(num_iris.corr(),annot=True, cmap='cubehelix', linewidths=1,linecolor='k',square=True,mask=False,vmin=1,vmax=1,cbar_kws={'orientation':'vertical'},cbar=True)
```

#### Distribution plot

```python
iris.hist(edgecolor='black',linewidth=1.2)
fig= plt.gcf()
fig.set_size_inches(12,6) # histogram for each attribute in the dataframe
```

#### Swarm plot

```python
sns.set(style='darkgrid')
fig= plt.gcf()
fig.set_size_inches(10,7)  # it is similar to the strip plot the main difference is each point remains distinct overr the data.
fig=sns.swarmplot(x='Species',y='PetalLengthCm',data= iris,hue= "Species")
```

```python
sns.set(style='whitegrid') # combinig the violin plot and swarmplot
fig= plt.gcf()
fig.set_size_inches(10,7)
ax=sns.violinplot(x='Species',y='PetalLengthCm',data= iris,hue= "Species", inner= None)
ax=sns.swarmplot(x='Species',y='PetalLengthCm',data= iris,color='black', edgecolor='white')
```

#### boxen plot

#### kde plot

```python
sub=iris[iris['Species']=='Iris-setosa']
sns.kdeplot(data=sub,x='SepalLengthCm',y='SepalWidthCm',cmap='plasma',shade = True, shade_lowest=False)
plt.title('Iris-setosa')
plt.xlabel('Sepal Length Cm')
plt.ylabel('Sepal Width Cm')
plt.show()
```

#### Stacked Histogram

```python
iris['Species'] = iris['Species'].astype('category')
```

```python
list1=list()
mylables= list()
for gen in iris.Species.cat.categories:
    list1.append(iris[iris.Species==gen].SepalLengthCm)
    mylables.append(gen)

h=plt.hist(list1,bins=30,stacked=True,rwidth=1,label=mylables)
plt.legend()
plt.show()
```

#### Area plot

```python
iris.plot.area(y=['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm'],alpha=0.4,figsize=(12, 6));
plt.show() #Areaplot displays the magnitude of change over time or across categories by filling the area
```
