```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,6))
```

```python
#1. Scatterplot
sns.scatterplot(x='total_bill', y='tip',data=tips, hue='time', size='size', palette='dark')
```

```python
# 2.line plot

sns.lineplot(data= tips, x='size', y='total_bill', hue='sex', markers='0' )
plt.title('Lineplot of Total bill Vs Size')
plt.show
```

```python
# line plot - ci =none

sns.lineplot(data= tips, x='size', y='total_bill', hue='sex', ci=None,markers='0' )
plt.title('Lineplot of Total bill Vs Size')
plt.show
```

```python
tips.columns
```

```python
# 3.barplot

sns.barplot(x='day', y='total_bill', data= tips, hue='sex',palette='muted')
plt. title('Bar plot of total bill by day')
plt.show
```

```python
# 4.box plot

sns.boxplot(data= tips, x='day', y='tip', hue='smoker', palette='muted')
plt.title('Boxplot of tips by Day and Smoker Status')
plt.show()
```

```python
# 5. Violinplot
sns.violinplot(data=tips, x='day', y='total_bill', hue='time', split= True,palette='pastel')
plt.title('Violinplot of Total Bill by Day and time')
plt.show()
```

```python
 #6. count plot 

sns.countplot(data= tips, x='day', hue='smoker',palette='dark')
plt.title('Count plot of Days by Smoker Stutus')
plt.show()
```

```python
# 7.  regression plot
sns.regplot(data= tips, x='total_bill', y='tip',scatter_kws={'s':50},line_kws={'color':'red'})
plt.title('Regression plot of total Bill vs Tip')
plt.show()
```

```python
#8. Histgram of total bill with KDE

sns.histplot(data= tips,x='total_bill',bins=20, kde=True, color='blue')
plt.title('Histgram of total bill with KDE')
plt.show()
```

```python
#9.pair plot

sns.pairplot(tips,hue='sex',vars=['total_bill','tip','size'],palette='husl')
plt.suptitle('Pair PLot: numeric variables by gender', y=1.02)
plt.show()
```

```python
 #10. catplot
sns.catplot(data=tips, x='day', y='tip', hue='sex', kind='point', palette='bright')
plt.title('catplot(point): Tips by day and gender')
plt.show()
```

```python
#11. jointplot

sns.jointplot(data=tips, x='total_bill', y='tip', kind='scatter',hue='smoker',palette='coolwarm')
plt.suptitle('Jointplot: Total Bill vs Tip',y=1.02)
plt.show()
```

```python
# 12. Facetgrid

g=sns.FacetGrid(tips, col='time', row='smoker',margin_titles=True).map(sns.histplot, 'total_bill',bins=20, kde=True)
g
```

```python
# 13. strip plot

sns.stripplot(data=tips, x='day', y='tip', hue='sex', jitter= True, palette='Set1')
plt.title('Strip Plot: Tips by data and gender')
plt.show()
```

```python
#14. KDE plot

sns.kdeplot(data= tips, x='total_bill', hue='sex', fill=True, palette= 'tab10')
plt.title('KDE Plot: Total Bill density by gender')
plt.show()
```
